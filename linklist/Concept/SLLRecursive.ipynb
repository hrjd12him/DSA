# Insert a node recursively

```python
class node():
    def __init__(self,data):
        self.data = data
        self.next = None


def insert_recursive(head,data,i):
    if i<0:
        return head
    
    if i == 0:
        newnode = node(data)
        newnode.next = head
        return newnode
    
    if head is None:
        return None
    
    smallhead = insert_recursive(head.next,data,i-1)
    head.next = smallhead
    return head

def printll(head):
    while head is not None:
        print(str(head.data)+"->",end="")
        head = head.next
    print("None")

def takeinput():
    head = None
    tail = None
    inputlist = [int(ele) for ele in input("enter you list input: ").split()]
    for currvalue in inputlist:
        if currvalue == -1:
            break

        newnode = node(currvalue)
        if head is None:
            head = newnode
            tail = newnode
        else:
            tail.next = newnode
            tail = newnode

    return head

head = takeinput()
printll(head)
no = int(input("how many time you need to insert a data: "))
while no>0:
    data = int(input("Enter the data: "))
    i = int(input("Enter the ith position: "))
    insert_recursive(head,data,i)
    printll(head)
    no-=1
else:
    print("The final linklist is :" ,end="")
    printll(head)
```

# Deleting a node recursively 

```python
class node():
    def __init__(self,data):
        self.data = data
        self.next = None

def DDLL_recusively(head,i):

    if i<0 or head is None:
        return head
    
    if i == 0:
        head = head.next
        return head     

    smallhead = DDLL_recusively(head.next,i-1)
    head.next = smallhead
    return head

def printll(head):
    while head is not None:
        print(str(head.data)+"->",end="")
        head = head.next
    print("None")

def takeinput():
    head = None
    tail = None
    inputlist = [int(ele) for ele in input("enter you list input: ").split()]
    for currvalue in inputlist:
        if currvalue == -1:
            break

        newnode = node(currvalue)
        if head is None:
            head = newnode
            tail = newnode
        else:
            tail.next = newnode
            tail = newnode

    return head

head = takeinput()
printll(head)
i = int(input("enter the index position for deleting a node: "))
DDLL_recusively(head,i)
printll(head)
```
