```python
class BinaryTreeNode():
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right =None

class BST:
    def __init__(self):
        self.root = None
        self.Nonodes = 0

    def printTreeHelper(self,root):
        if root == None:
            return 
        print(root.data,end=':')
        if root.left != None:
            print('L:',root.left.data,end=',')
        if root.left == None:
            print('L:',None,end=' ,')
        if root.right != None:
            print('R:',root.right.data,end = '')
        if root.right == None:
            print('R:',None,end=' ')
        print()
        self.printTreeHelper(root.left)
        self.printTreeHelper(root.right)


    def printTree(self,root):
        return self.printTreeHelper(self.root) 
    
    def isPresenthelper(self,root,data):
        if root == None:
            return 
        if root.data == data:
            return True
        if root.data>data:
            return self.isPresenthelper(root.left,data)
        else:
            return self.isPresenthelper(root.right,data)
            
    def isPresent(self,data):
        return self.isPresenthelper(self.root,data)
    
    def inserthelper(self,root,data):
        if root ==None:
            node = BinaryTreeNode(data)
            return node
        if root.data>data:
            root.left = self.inserthelper(root.left,data)
            return root
        else:
            root.right = self.inserthelper(root.right,data)
            return  root
    
    def insertdata(self,data):
        self.Nonodes+=1
        self.root = self.inserthelper(self.root,data)
    
    def mindata(self,root):
        if root == None:
            return 100000
        if root.left == None:
            return root.data
        
        return self.mindata(root.left)
        
    def deletehelper(self,root,data):
        if root  == None:
            return False, root
        
        if root.data<data:
            deleted,newrightroot =self.deletehelper(root.right,data)
            root.right =newrightroot
            return deleted,root
        
        if root.data>data:
            deleted,newleftnode =self.deletehelper(root.left,data)
            root.left = newleftnode
            return deleted,root
        
        if root.left ==None and root.right ==None:
            return True,None
        
        if root.left == None:
            return True,root.right
        
        if root.right ==None:
            return True,root.left
        
        replacement =self.mindata(root.left)
        root.data = replacement
        deleted ,newrightroot=self.deletehelper(root.right,replacement)
        root.right =newrightroot
        return True,root

    def deletedata(self,data):
        deleted,newroot = self.deletehelper(self.root,data)
        if deleted:
            self.Nonodes-=1
        self.root = newroot
        return deleted


    def count(self):
        return self.Nonodes
    
b = BST()
root = b.insertdata(10)
b.insertdata(20)
b.insertdata(8)
b.insertdata(15)
b.insertdata(7)
b.insertdata(9)
b.insertdata(21)
b.printTree(root)
print()
print(b.deletedata(21))
print()
b.printTree(root)
```
