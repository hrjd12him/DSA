```python
import queue
class treenode():
    def __init__(self,data):
        self.data = data
        self.left =None
        self.right = None

def inputlevelwise():
    q = queue.Queue()
    print('enter root data: ')
    rootdata = int(input())
    if rootdata ==-1:
        return
    root = treenode(rootdata)
    q.put(root)
    while not q.empty():
        curr_node = q.get()
        print('Enter the left Child of node',curr_node.data)
        leftchilddata =int(input())
        if leftchilddata != -1:
            leftchild = treenode(leftchilddata)
            q.put(leftchild)
            curr_node.left = leftchild
        print('Enter the right child of node',curr_node.data)
        rightchilddata = int(input())
        if rightchilddata != -1:
            rightchild = treenode(rightchilddata)
            q.put(rightchild)
            curr_node.right = rightchild
    return root

def printlevelorder(root):
    if root ==None:
        return 
    q =[]
    q.append(root)
    while q:
        count = len(q)
        while count>0:
            curr_node =q.pop(0)
            print(curr_node.data,end=' ')
            if curr_node.left:
                q.append(curr_node.left)
            if curr_node.right:
                q.append(curr_node.right)

            count-=1
        print(' ')

def btwk1k2(root,k1,k2):
    if root == None:
        return
    if root.data<k1:
        btwk1k2(root.right,k1,k2)
    elif root.data>k2:
        btwk1k2(root.left,k1,k2)
    else:
        print(root.data,end=' ')
        btwk1k2(root.left,k1,k2)
        btwk1k2(root.right,k1,k2)

    

root =inputlevelwise()
printlevelorder(root)
k1 = int(input('Enter the starting range value: '))
k2 = int(input('Enter the Ending range value: '))
btwk1k2(root,k1,k2)
```
