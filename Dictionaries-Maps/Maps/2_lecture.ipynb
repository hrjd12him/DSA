````python
class mapnode:
    def __init__(self,key,value):
        self.key = key
        self.value = value
        self.next = None

class Map:
    def __init__(self):
        self.bucketsize = 10
        self.bucketarray = [None]*(self.bucketsize)
        self.count = 0


    def size(self):
        return self.count
    

    def findindex(self,hc):
        return abs(hc)%(self.bucketsize)
    

    def insert(self,key,value):
        hc = hash(key)
        index = self.findindex(hc)
        head = self.bucketarray[index]
        while head is not None:
            if head.key ==key:
                head.value = value
                return 
            head = head.next
        head=self.bucketarray[index]
        newnode= mapnode(key,value)
        newnode.next = head
        self.bucketarray[index] = newnode
        self.count +=1

    def remove(self,key):
        hc = hash(key)
        index = self.findindex(hc)
        head = self.bucketarray[index]
        prev = None
        while head is not None:
            if head.key == key:
                if head.next == None:
                    self.bucketarray[index] = head.next
                else:
                    prev.next = head.next
                self.count-=1
                return head.value
            prev = head
            head = head.next
        return None

    def getvalue(self,key):
        hc = hash(key)
        index = self.findindex(hc)
        head = self.bucketarray[index]
        while head is not None:
            if head.key == key:
                return head.value
            head = head.next
        return None
    
d= Map()
d.insert('abc',10)
d.insert('abc',20)
d.insert('xyz',1)
d.insert('abe',99)
d.insert('acl',32)
d.size()
print(d.remove('abe'))
d.size()

````
